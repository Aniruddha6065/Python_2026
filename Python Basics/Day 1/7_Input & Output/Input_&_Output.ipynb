### Taking Input using input()

<small>Python's input() function is used to take user input. By default, it returns the user input in form of a string.</small>

```python
name = input("Enter your name: ")
print("Hello,", name, "Welcome!")

''' The code prompts the user to input their name, stores it in the 
variable "name" and then prints a greeting message addressing the user 
by Atheir entered name.'''
```
